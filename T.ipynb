### 1. Importamos paqueteria pandas

```python
import pandas as pd
```

### 2. Cargamos base de datos y seleccionamos variables de interes: Portafolio, iniciativa, nombre, Owner, administración general, fecha de inicio, fecha de fin, estado

```python
portafolio = pd.read_excel('portafolio_completo.xlsx', sheet_name='Portafolio completo')
portafolio = portafolio.loc[:,['Portafolio','Iniciativa','Nombre','Owner', 'Start', 'Finish','Administración General','Estado']]
portafolio
```

### 3. Creamos funcion para dar formato deseado a la fecha

```python
def date_format(column):
    
    meses = {"/01/":"Enero", "/02/":"Febrero",
         "/03/":"Marzo",  "/04/":"Abril",
         "/05/":"Mayo",  "/06/":"Junio", 
         "/07/":"Julio", "/08/":"Agosto",
         "/09/":"Septiembre", "/10/":"Octubre",
         "/11/":"Noviembre", "/12/":"Diciembre"}
    
    for i in range(len(column)):
        dia = column.iloc[i][0:2]
        mes = column.iloc[i][2:6]
        año = column.iloc[i][6:10]
        column.iloc[i] = mes + ' ' + dia + ", " + año

    column = column.replace(meses, regex=True)
    return column
date_format(portafolio['Start'].dropna())
```

### 4. Eliminamos los poyectos que no tengan una fecha de inicio ni una fecha de fin

```python
portafolio['Administración General'].value_counts()
```

```python
portafolio = portafolio.dropna(subset = ["Start"])
portafolio['Start']
```

### 5. Utilizamos función que creamos en punto 4 para dar el formato deseado a las fechas 

```python
portafolio['Start'] = date_format(portafolio['Start'])
portafolio['Finish'] = date_format(portafolio['Finish'])
portafolio
```

### 6. Renombramos las Administraciones generales y estado para obtener formato deseado

```python
portafolio = portafolio.replace({'AGA':'solid circle', 'AGSC':'solid triangle', 'AGCTI':'solid star', 'AGR':'solid play','AGAFF':'solid diamond', 'AGRS':'solid bold', 'AGJ': 'solid cloud', 'AGGC': 'solid heart', 'AGP': 'solid sun', 'AGACE': 'solid lightbulb', 'AGE': 'solid megaphone'})
portafolio['Administración General'].value_counts()
```

```python
portafolio = portafolio.replace({'Activo':'#19CCD2', 'Detenido con CCRS': '#D2194A', 'Detenido': '#D71995'})
portafolio['Estado'].value_counts()
```

```python
portafolio = portafolio.rename(columns={'Portafolio': 'layer_name', 'Iniciativa': 'event_id', 'Nombre': 'event_name', 'Owner': 'notes', 'Start': 'start_date', 'Finish': 'end_date', 'Estado': 'color', 'Administración General': 'icon'})
portafolio
```
